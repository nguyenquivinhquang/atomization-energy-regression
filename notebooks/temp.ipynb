```python
def mean_variance(array):
    mean = np.mean(array)
    var = np.sum((array - mean)**2) / (len(array))
    std = np.sqrt(var)
    return mean, var
```

```python
dis = []
label = []
for mol in mols_infor:
    if len(mols_infor[mol])< 3: continue
    mean, var = mean_variance(mols_infor[mol])
    std = np.sqrt(var)
    # if std < 40: continue
    dis.append(mean)
    label.append(mol)
    print(mol, mean, var, std, mols_infor[mol])
```
